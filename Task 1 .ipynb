# мое решение

```python
n, k = map(int, input().split())
def Cnk(n, k):
    if k == 0:
        return 1
    elif k > n:
        return 0
    else:
        return Cnk(n-1, k) + Cnk(n - 1, k - 1)
print(Cnk(n, k))
```

# #Решение учителя

```python
#1

def c(n, k):
    if k > n:
        return 0
    if k == 0:
        return 1
    return c(n - 1, k) + c(n - 1, k - 1)

n, k = map(int, input().split())
print(c(n, k))
```

```python
2#

n, k = map(int, input().split())

sz = max(n, k) + 1
c = [[0] * sz for _ in range(sz)]

c[0][0] = 1
for i in range(1, sz):
    for j in range(i + 1):
        c[i][j] = c[i - 1][j] + c[i - 1][j - 1]

print(c[n][k])
```
